The funerary epigraphy of Thugga (Dougga), Roman North Africa, is characterised by an unusual reliance on a limited range of formulae. This research investigates to the extent that this is statistically real by applying a corpus linguistic methodology to assess standardisation metrics at Thugga, Uchi Maius, Thibursicum Bure, Lambaesis and Rome. This comparative analysis involves 8,002 funerary inscriptions drawn from the Epigraphik-Datenbank Clauss-Slaby (EDCS) and compiled with a custom-built Python tool. A corpus-lingustic analysis of this data points to the existence of distinct funerary record at Thugga, potentially related to a local interpretation of the Manes and a divergence in values perceived as worth commemorating. In doing so, this research sheds light on the complex and localised religious environment of Roman North Africa, in which pre-existing Numidian chthonic traditions and incoming Roman spiritual practices were actively renegotiated rather than simply replaced or assimilated. 

## Introduction

North Africa and the Maghreb possess a rich archaeological record, and the study of epigraphy has dominated scholarship in the region (Mattingly and Hitchner, 1995:166-169). The region has long existed as a cultural crossroad, which is reflected in its tradition of diverse commemorative practices (Stone and Stirling, 2007:7). The transformation of the region into a Roman province was iterative, beginning after the destruction of Carthage and annexation of Punic territory in 146 BCE, a border being delineated between the new Roman province of Africa Vetus and the Numidian Kingdom 90km inland (Hobson, 2022:103). By 43 BCE, the remaining Numidian lands were formally annexed into the growing Empire. This year also saw the refoundation of Carthage with Roman colonists, alongside the arrival of Roman settler communities to many pre-existing towns (Hobson, 2022:103). One such town was Thugga, a former Numidian centre and presently one of the best-preserved examples of a small Roman town in North Africa (UNESCO, 2026). It is host to an extensive epigraphic record, with over 2000 funerary inscriptions from Thugga digitised on the online repository Epigraphik-Datenbank Clauss-Slaby (EDCS). A browse of this repository reveals a recurring pattern: the same small cluster of abbreviations, a name, and an age at death. The inscriptions are extremely short, deploy virtually identical formulae, and are devoid of any reference to profession, personal narrative, or elaboration. Particular phrases are unusually dominant, in particular those relating to "*pietas*".

Although this is not necessarily unusual in Roman funerary epigraphy (Hope, 2009:166) , Thugga has been identified as particularly constrained outlier before. Bennett identified Thugga's anomalous reliance on a restricted formulaic register in a macro-level analysis of over 100,000 Roman epitaphs as part of a PhD thesis, attributing it to copybook stonemasons, but explicitly called for more detailed future research (Bennett, 2020:194). This research intends to compare Thugga to comparable sites to determine to what extent it is an outlier, and specifically in which way, before proposing a possible hypothesis as to why this is the case. To do so it employ corpus-lingustics. Corpus linguistics is a quantitative methodology used within the field of linguistics, in which computer software is used to identify and count a wide range of linguistic features, often with the objective of producing statistical models (McEnery and Wilson, 2001). It can be applied to many aspects of language and therefore to funerary epigraphy, measuring anything from the prevalence of certain formulae to broader patterns of collocation or the extraction of key terms. A prerequisite is a dataset of machine-readable text, the assembly of which is logistically difficult and a significant constraint to conducting macro-level epigraphic research of this kind, typically requiring painstaking manual labour or the production of custom-built software tailored to the project (Hardie, 2016:510). This creates a technical barrier, as the methodology section demonstrates, though one now surmountable through large language model-assisted development. Corpus-linguistics has always being intimately tied to the humanities, with the first corpus being produced by the Jesuit priest and PhD student Roberto Busa whilst studying the writings of Thomas Aquinas, a project completed in 1967 (McEnery and Wilson, 2001:20). Although corpus linguistics has been used in the study of knowledge production in archaeology (Schwartz, 2024), large-scale comparative studies applied directly to epigraphic data nonetheless remain rare (Hermankova et al, 2022).


## Methadology 

Corpus linguistics remains largely restricted to the discipline of linguistics itself. Although a recent paper, published in March of 2026 assessed a corpus of 1650 late Roman inscriptions using corpus linguistics, it did so through an explicitly sociolinguistic, rather than archaeological lens. Specifically, to assess the reduction of AE to A in Latin inscriptions between the fourth and sixth centuries (Papini, 2026:112) something associated with the later evolution of Latin into a proto-Romance language. Bennett’s work employed a similar methodology, however largely depending on excel regex searches rather than corpus linguistic software, to identify common formula and assess other metrics such as average characters per inscription (Bennett, 2020:64). However, Bennett’s extremely broad approach meant that beyond calculating average inscription length and identifying prevalence of certain formula via regex searches, a more detailed analysis was not possible.

This study addresses that gap by applying dedicated corpus linguistic software to a smaller, targeted set of five comparable sites, enabling a more detailed methodological approach than Bennett's macro-level survey allowed. Despite this narrower remit, this research includes nearly the entire digitised preservation record from Thugga, Lambaesis Uchi Maius and Thiburiscium Bure. Following site selection, multiple metrics were assessed. First, standardisation metrics were used to measure lexical diversity across each corpus. Second, key terms relating to social and professional identity were extracted from each site's corpus, and the most frequently occurring terms compared across sites. Finally to visualise how formulaic construction was assessed, a template tree was constructed from smaller corpora, mapping the branching sequence of dominant grammatical/formulaic elements within the corpus from Thugga compared to another site. 


## Site Selection and Dating

### Thugga

Thugga is central to this research, previously identified as a possessing an unusual epigraphic habit. Not only does it possess an extremely rich and well-preserved epigraphic record, but also an interesting and unusual cultural history. The date of the foundation of Thugga is unknown, but it was first recorded in the 4th century BCE when it was likely largely self-governing (Fentress and Dufton, 2022:191). It was not conquered violently, but shortly after the annexation of Numidian lands Roman settlers arrived, establishing a *Pagus* (Hitchner, 2002:598). This led to the development of an unusual dual administrative system defined by the existence of *Sufetes* (Fentress and Dufton, 2022:191) operating in parallel to the Roman magistrates and *Pagus*, for over two centuries (Hitchner, 2002:599).

### Lambaesis 

Lambaesis was first founded as a Roman military fort in 81 CE, and quickly also developed into a civilian town (Wilson, 2016). It would become the headquarters for Legio III Augusta, the only legion stationed in the region and one that would eventually be staffed largely by local recruits by the time of the second century (Faure, 2022:164). Lambaesis like Thugga has an excellent preservation record, with over 1500 individual funerary inscriptions digitised on EDCS. Additionally, it was host to a mix of cultural influences. Initially, the garrison comprised foreign soldiers alongside local civilians, many of whom intermarried. Over time it became host to soldiers of foreign descent born in Africa, as well as a large number of locals. The multi-ethnic nature of the Roman Army in North Africa “cannot be doubted” (Faure, 2022 p.164). This makes it an interesting candidate for corpus linguistic research, especially in comparison to Thugga, a city exposed to significantly different influences.

### Rome

Rome was selected as a metropolitan “baseline”. I wished to compare the data from Lambaesis and Thugga, with a city from outside of North Africa. This was necessary to differentiate between trends local to North Africa, and the individual site. Rome was the most obvious choice for this, as an imperial capital with the most expansive epigraphic record in the Roman world, making it ideal for corpus creation.

### Thuburiscium Bure and Uchi Maius 

These are both smaller sites, located in close proximity to Thugga. They were selected as additional comparisons to assess whether patterns observed at Thugga extended to neighbouring communities as observed by Bennett, 2020:189.

## EDCS CorpusBuilder

The assembly of these corpora required the creation of bespoke software - CorpusBuilder which is a custom Python tool built around a text normalisation script. It strips unnecessary EDCS metadata and eliminates author expansions identifiable by their appearance within () and <> brackets while preserving restored letters within []. It identifies 29 common abbreviations frequent in Latin funerary epigraphy (Elliot, 1998), capitalises all text, and converts U to V throughout. The program consists of 1881 lines of code and can generate funerary corpora from 10 sites, all North African except Rome, using source data extracted from EDCS in March 2025. Corpus generation is controlled through a Tkinter GUI, through which the researcher selects the data source, number of inscriptions, date range, inclusion or exclusion of Greek texts, and a damage threshold of inscriptions with 3 eligible characters or fewer are excluded. Once the filters specified by the key have been applied and non-conforming inscriptions eliminated from the candidate pool, the SHA-256 calculation is performed. SHA-256 assigns a unique hash first to the dataset as a whole, then to every individual EDCS ID within it. The final input to this calculation is the generated key itself.

$$\text{Score} = \text{SHA256}(\text{key} \mathbin{\|} \text{dataset hash} \mathbin{\|} \text{EDCS ID})$$

Entering a key therefore amounts to supplying identical values to this equation, which necessarily yields identical results, and consequently the same ranking of inscriptions, and thus the same reproduced corpus, both in terms of which inscriptions are selected and the order in which they appear. SHA-256 is designed as a cryptographic tool and was selected primarily due its extremely low collision rate (virtually 0 or to be specific 1 in 2²⁵⁶) when compared to non-cryptographic hash alternatives. An added bonus is that every corpus generation ultimately consists of randomly selected inscriptions with the caveat of them confirming to filtering selection. 


```python
def _sha256_file(p: Path) -> str:
    h = hashlib.sha256()
    with p.open("rb") as f:
        for chunk in iter(lambda: f.read(1 << 16), b""):
            h.update(chunk)
    return h.hexdigest()


def dataset_fingerprint(files: List[Path]) -> str:
    pairs = []
    for p in sorted([Path(x) for x in files], key=lambda x: x.name.lower()):
        try:
            pairs.append((p.name, _sha256_file(p)))
        except Exception:
            pairs.append((p.name, "ERROR"))
    h = hashlib.sha256()
    for name, sh in pairs:
        h.update(name.encode("utf-8"))
        h.update(b"\0")
        h.update(sh.encode("ascii", errors="ignore"))
        h.update(b"\n")
    return h.hexdigest()


def deterministic_select(
    inscriptions: List[Tuple[str, str]],
    n: int,
    seed_canon: str,
    ds_fp: str,
) -> List[Tuple[str, str]]:
    if n <= 0:
        return []

    def score(eid: str) -> bytes:
        msg = f"{seed_canon}|{ds_fp}|{eid}".encode("utf-8")
        return hashlib.sha256(msg).digest()

    ranked = sorted(inscriptions, key=lambda it: (score(it[0]), it[0]))
    return ranked[:n]
```

### Program Limitations
When using two corpora from the same site, there is a chance that they will contain overlapping inscriptions. This not necessarily a limitation but should be noted. The only way to confirm overlap would be to manually cross reference EDCS: IDs. This would be time consuming if done manually, but with the help of a corpus-linguistic tool/software it can be completed relatively easily. A further limitation is that it is not possible to modify the parameters of an already-generated corpus while retaining the same key. Any change to a filter or setting after generation triggers the invalidation system, which clears the stored seed and assigns a new one on the next build. Reproducing a prior corpus with altered settings would therefore require manually reconstructing the key token with the desired parameters. This process while straightforward given the key's readable format, is not currently supported within the GUI itself. Finally the tool has not undergone extensive bug testing, and has only being used for to generate the corpora required for this research.

### Dating
Accurate dating is difficult, as even inscriptions with dating metadata typically consist of broad approximations rather than specific dates, sometimes spanning four centuries (Hansen, 2021: 142). An initial attempt to construct century-by-century corpora proved unworkable for this reason, as the resulting datasets were nearly identical. Site history is therefore used as a proxy for estimated date ranges. At Thugga urban space began contracting by the fourth century, and by the seventh (Von Rummel, 2016: 109). The Thugga corpus is accordingly assigned a first-to-fourth century date range, with the caveat that funerary inscription production likely peaked alongside general construction activity in the second and early third centuries (Cilliers, 2016:88). Lambaesis follows a comparable trajectory, effectively collapsing by the fifth century, with parts of the urban area converted to grave use as early as 350 AD (Roskam, 1996: 47). Both North African corpora are therefore broadly chronologically analogous, consisting predominantly of early and middle imperial material. This estimate is extended to both Uchi Maius and Thiburiscium Bure, both located within 12km of Thugga. To ensure comparability, the Roman corpus was restricted to inscriptions dated between 1–400 AD, which the higher proportion of precisely dated Roman inscriptions makes possible.

## Measuring Standardisation and Formulaic Patterning 
Measuring lexical diversity is frequently an objective of corpus-linguistic study (Tweedie and Baayen, 1998:323), which makes it particularly applicable to Thugga and the assessment of its seemingly highly constrained funerary epigraphy. There are multiple metrics that are used to assess a corpus’s standardisation. One of the most frequently used and mathematically simple ones is the type-token ratio (TTR). The TTR of each site is calculated by dividing the total number of types (unique words/abbreviations) by total number of tokens (words/abbreviations). This identifies the percentage of unique vocabulary that appears in each corpus. It is important to note that funerary epigraphy is inherently more constrained than typical linguistic datasets that consist of free-flowing text. It is also important to note that TTR values are suppressed in proportion to corpus size (Tweedie and Baayen, 1998:323). TTR can be summarized as a measurement of the proportion of unique words, and as new words cannot be introduced faster than total number of words larger corpora will always return lower TTR scores. In this instance Rome, as the largest corpus will return a TTR score that is disproportionately lower (appearing more standardized) than either Lambaesis or Thugga. 

This is referred to as the corpus size effect. A solution to this is the implementation of Herdan’s C TTR, a logarithmic modification to the TTR calculation that accounts for the corpus size effect. Ultimately it measures the rate at which new words (types) are added in proportion to the total number of words (tokens), rather than just measuring the unique vocabulary of a corpus. It was implemented as a further verification of TTR results. The average number of tokens per inscription were also calculated, to complement the findings of Bennett who calculated the average number of characters per inscription. Additionally a custom index was devised, the formulaic concentration index or FCI which measures the proportion of the entire corpus that consists of the top 10 most frequently appearing terms. This is an empirical way to measure a corpus’s reliance on specific limited formulae, rather than a general metric of standardization. 

Template maps were constructed to map the dominant formulaic structure of Thugga's funerary epigraphy. Using AntConc's KWIC tool, the frequency with which inscriptions followed each successive formulaic element was calculated as a relative percentage, with each branch using the preceding term as its denominator. For example, of the 87.27% of inscriptions beginning with DMS, 100% follow this with the name of the deceased. The resulting trees map the corpus's dominant formulaic patterns rather than accounting for every inscription. KWIC searches were additionally used to identify key terms and abbreviations relating to social and professional roles, including familial identifiers, which were exported and compared in tabular form. All structural analysis was conducted on the two smaller dated corpora from Thugga and Lambaesis. The Lambaesis corpus has a smaller date range (100-200CE). This was necessary to reduce its size, as it was otherwise almost impossible to construct a lexical distribution tree due to the diversity of terms and order. To extract personal identifiers AntConc’s tag system was used, searching for a pre-loaded list of 63 relevant terms and abbreviations in both corpora. 


### Limitations
Dating uncertainty is the most significant limitation. Site history implies the corpora largely consist of first-to-fourth century inscriptions, but a narrower objective date range cannot be assigned, making it difficult to detect chronological variation within the datasets. A small number of inscriptions likely fall outside this range entirely, and while Thugga and Lambaesis appear to have followed similar trajectories of decline, the corpora cannot be confirmed as precisely chronologically equivalent. The 2 smaller datasets from Lambaesis and Thugga are of differing date ranges albeit they overlap, Lambaesis occupying a much narrower range of 100-200 CE compared to Thugga’s 1-400 CE, a necessary constraint for creating a lexical distribution tree at Lambaesis altogether. 

Abbreviation tokenisation posed a methodological challenge. Without intervention, AntConc tokenises abbreviations such as DM or D.M into single-letter tokens, inflating the total token count and rendering output uninterpretable. CorpusBuilder was designed with a list of common funerary abbreviations to be merged into single lexical units, initially comprising 12 entries. However, single-letter tokens continued to appear among the top ten most frequent terms at each site, with Rome a notable outlier. Expanding the list from 12 to 29 abbreviations produced only a 0.78% reduction in Rome's token count (310 tokens), indicating that the majority of single-letter tokens at Rome are genuine initials rather than split abbreviations, and were not significantly distorting the results. It must also be noted that EDCS inscription counts do not correspond exactly to individual epitaphs. EDCS counts individual stones, several of which contain multiple epitaphs typically of family members. This is why the Thugga dated corpus of 233 inscriptions actually contains 275 individual epitaphs. This discrepancy does not critically affect the analysis but should be noted.
 

# Results
All standardization metrics reveal a consistent hierarchy of uniformity. Thugga demonstrates the most restricted vocabulary, followed by Lambaesis, with Rome exhibiting the greatest lexical diversity. This same pattern also appears in the much smaller, dated corpora but with smaller calculated difference.
The dated corpora (which consist of just a combined 347 inscriptions) were used for more granular analysis. Alongside this, terms pertaining to professional and social roles were extracted using the KWIC function in Antconc. These corpora were also used as the basis for the creation of lexical distribution trees. 
The undated corpora (which consist of a combined 6128 individual inscriptions) were used for macro-analysis and the quantitative measurement of linguistic features and standardization between the three sites. This included the extraction of the 100 most frequent terms appearing in each site, a necessity to perform the Formulaic Concentration Index calculations. The total number of lexical units (tokens) and number of unique lexical units (types) were also extracted and used to calculate TTR and Herdans C scores.  


## Average Epitaph Length
The average length of inscriptions was calculated by dividing the total number of tokens (words) per corpus (N)  (measured in Antconc as total frequency) by the total number of inscriptions in each corpus . 

$$\text{Average Inscription Length} = \frac{\text{Tokens} (N)}{\text{Inscriptions} (I)}$$

| Site | Tokens (N) | Inscriptions (I) | Average Inscription Length |
|------|-----------:|------------------:|----------------------------:|
| Thugga | 20,628 | 2,048 | 10.0 |
| Lambaesis | 25,009 | 1,585 | 15.8 |
| Rome | 39,664 | 2,000 | 19.8 | 

Thugga emerges as a significant outlier, Roman epitaphs average nearly twice as long (98% longer), while inscriptions from Lambaesis exceed Thugga's by 57.7%. Interestingly the gap between Lambaesis and Rome is less significant, than that of Thugga and Lambaesis despite geographic proximity to each other. These differences, detected between such large corpora allude to structural differences in commemorative practice.

## Type Token Ratio
TTR score was calculated by extracting unique lexical types per corpus, measured in antconc as “entries” (V), which was then divided by tokens (N). TTR is a standard measure of lexical diversity, however TTR values inversely correlate to corpus size, meaning larger corpora will always appear to be less lexically diverse when compared to smaller ones (Lisson et al, 2018). As we are comparing corpora of different sizes, Rome will record a disproportionately lower TTR score due to corpus size effects, rather than real standardization (Mitchell, 2014). Equally Lambaesis as the smallest corpus will yield a comparatively higher TTR score. These results were presented to three decimal places, as smaller mathematical differences are representative of significant real-world differences. 

$$TTR = \frac{\text{Types} (V)}{\text{Tokens} (N)}$$

| Site | Types (V) | Tokens (N) | TTR |
|------|----------:|-----------:|:---:|
| Thugga | 2,650 | 20,628 | 0.128 |
| Lambaesis | 3,661 | 25,009 | 0.146 |
| Rome | 7,077 | 39,664 | 0.178 |

Despite mechanical compression of TTR scores, site ranking remains unchanged, however Lambaesis is closer to Thugga then Rome. Rome is a significant outlier. As the largest corpus Rome experiences the most mechanical suppression of its TTR score, despite this it maintains a 0.05 (39%) and 0.032 (22%) point higher score than Thugga and Lambaesis respectively. These results require qualification, as comparing corpora of differing sizes is inherently problematic because of artificial score compression. Regardless, the hierarchy first observed in average inscription length is preserved.

## Herdans C
A modification to the TTR calculation was devised by the linguist Gustav Herdan to empirically account for TTR score compression, referred to as Herdans C. (Mitchell, 2014). It does this by introducing the logarithmic function to the TTR calculation, measuring the number of unique words added to the corpus in proportion to the number of total words. In short, it measures lexical diversity by measuring how quickly vocabulary grows as corpus size increases.

$$Herdan\ C = \frac{\log(V)}{\log(N)}$$

| Site | Types (V) | Tokens (N) | Herdan's C |
|------|----------:|-----------:|:----------:|
| Thugga | 2,650 | 20,628 | 0.79 |
| Lambaesis | 3,661 | 25,009 | 0.81 |
| Rome | 7,077 | 39,664 | 0.83 |

Even when accounting for corpus size, the same pattern of results appear. These results demonstrate that this pattern is representative of structural differences in funerary epigraphy between sites, rather than artificial compression of TTR values. However this time, the gap between sites is roughly even with a 0.02 gap between each site. Lambaesis produces longer epitaphs than Thugga (15.8 vs 10 tokens) but achieves this with somewhat constrained vocabulary.


## Formulaic Concentration Index
To further quantify standardization by site, it was necessary to create a custom index – the Formulaic Concentration Index (FCI). This measures the proportion of the whole corpus that consists of the top ten most frequent lexical types (words).

$$FCI = \frac{\text{Top 10 Tokens} (N_t)}{\text{Total Tokens} (N)} \times 100$$

| Site | Top 10 Tokens (N_t) | Total Tokens (N) | FCI |
|------|---------------------:|------------------:|:---:|
| Thugga | 5,559 | 20,628 | 27.0 |
| Lambaesis | 4,732 | 25,009 | 18.9 |
| Rome | 5,450 | 39,664 | 13.7 |

The FCI index results yield the most striking evidence yet for hyperstandardisation at Thugga. 27% of the entire corpus at Thugga consists of just 10 lexical units. This degree of formulaic concentration far exceeds both Lambaesis (18.9%) and Rome (13.7%). Lambaesis once again occupies an intermediate position, perhaps pointing to some shared commemorative preferences that are far more pronounced at Thugga.

# Discussion
When compared with Rome and Lambaesis, the funerary epigraphy of Thugga is highly unusual. Not only does it have a highly consistent and committed preference for short, standardized inscriptions, it does so by repeatedly invoking the same formulae. Additionally, any reference to the deceased’s professional identity (and largely social identity too) is almost entirely excluded. This is highly unusual, as the celebration of one’s conventional success was frequently a focus of Roman funerary epigraphy, the objective of which was to affirm personal memory (Hope, 2009:39). Military cemeteries are typically more standardized, as soldiers tend to reiterate a collective unit or military identity in death (Hope, 1997:116). This likely explains the disparity between Lambaesis and Rome. This, however, implies that Thugga is even more of an outlier, with both a lack of  individual epitaph variation (although it does exist) and a near total absence of any reference to the secular life of the deceased to the extent it is statistically significantly more standardized than a predominantly military site. 
Bennett argues that Thugga’s uniformity (and similar profiles found in nearby cities) may be the result of travelling stone masons with a reliance on copy books potentially mass-producing pre-cut stelae (Bennett 2020:192-211). This may be logistically true to some extent, it is not, however, a causal explanation. If we are to presume that Thugga’s epigraphic habit is a result of commercial convenience, we are also presuming that Thuggans did not care enough to try and commission more individual epitaphs and passively accepted what the market offered. In a somewhat similar vein to this, Shaw when referencing both the unusual degree of formulaic standardization in North Africa, and the absence of familial references argued that this was simply the result of poverty and space saving (Shaw, 1984:463). Both these arguments eliminate the agency of Roman North Africans and treats them as passive subjects of circumstance. I would argue that this view is not supported by the available evidence. 


## Commercial Necessity and Mass Production?
To assess the likelihood of the mass production hypothesis, a closer analysis is required. Although Bennett correctly noted the unusual dominance of certain formulae, and it is even possible to map out a dominant template, individual variation still exists, and the order in which these terms are presented is not always concrete. For example, although PVA (Pia Vixit Annis) is highly dominant, occurring 1047 times in a corpus of 2048 stones, P V AN, P V ANNIS, P VIX AN, PIVS VIXIT ANNIS, PIA VA, PIA VIXIT also appear and some variation/combination of this is what the other half of the corpus predominantly consists of.

Despite the fact that all these formulae are functionally almost identical, each one being some variation of “lived piously for x years”, the way in which this is invoked can differ significantly, and sometimes by as little as a single or a few letters (VIX vs VIXIT for example). Additionally, even of those that appear with PVA, there is often still some level of variation, for instance a number follow HSE with other abbreviations e.g “OTBQTTLS”. This diversity points to voluntary adherence rather than mass-production. It should also be noted, that as flagged by Bennett herself and as evidenced by multiple metrics assessed in this research, the preference for certain formula and standardization is not exclusive to Thugga (Bennett, 2020:156). This implies that we are not observing the result of pre-cut mass-produced stelae, but a deliberate commemorative preference that potentially existed throughout Africa Proconsularis but was strongest at Thugga and the immediate surrounding region.

A more developed version of the commercial necessity hypothesis is that a differing preservation environment at Thugga has privileged the preservation of shorter inscriptions produced by people from a lower socio-economic background. This is a real possibility as Thugga’s rapid decline and relative remoteness meant that only limited amounts of spolia were taken from the site which was left predominantly in-situ. Perhaps the corpora differences can be explained by the destruction of similar inscriptions in Rome which are simply less likely to survive extensive urban reuse. To test this hypothesis, a corpus of 2048 Roman inscriptions was generated with a length filter of 0-44 characters. This produced a corpus with an average length of 9.5, even shorter than Thugga’s 9.7. Herdens C and FCI scores were calculated and compared

$$Herdan\ C = \frac{\log(3157)}{\log(19573)} = 0.82$$

Remarkably even though the average inscription length is shorter than that of Thugga in this corpus, it still exhibits higher levels of lexical diversity, with the score now near identical to Lambaesis. This is also reflected by the short Corpus’s FCI score. 

$$FCI = \frac{2249}{19573} \times 100 = 11.5$$

The short Rome corpus exhibits extremely limited reliance its top 10 most frequently used terms, reducing from the other Roman corpuses score of 13. This is less them half of Thugga's score of 27. This is further confirmation that we are observing the product of cultural preference rather than one reflective of economic convenience or a preservation bias. In the short Roman corpus, DMS appears just 10 times, and pietas, and PVA none. 


## Ritual Adherence Beyond Secular Social Memory
Thugga hosted a unique pre-Roman religious enviroment. North Africa presented a patchwork of loosely related deities and place-based spirituality that did not constitute a codified religion, and this was not nullified by Roman annexation (Brett and Fentress, 1996:35-49). There appears to have also been a level of geographic divergence between the coastal belt and interior, defined by the dominance of chthonic cults in former Numidian lands (Mattingly, 2024:773) in which Thugga itself lies. Evidence for this chthonic tradition can be found in the city itself, which hosted a temple to the deified Numidian king Massinissa (McCarty, 2022:289). An outdated model of Romanization has frequently been applied to Thugga, particularly surrounding the arrival of the *Pagus*, but also when assessing its pre-Roman “Punic” or “African” character. For example, the existence of *Sufetes* at Thugga were assumed to be a preservation or direct adoption of the Carthaginian institution, in reality it appears to have been a local reinvention based on a distant memory with a significantly different set of functions and responsibilities (McCarty, 2017:412). Although it is true that the *Pagus* remodelled large parts of the forum and dismantled pre-existing temples (Fentress, 2021:965) there is evidence of a more complex religious relationship lurking beneath as well. 

For example, the people of Thugga continued to deposit stelae dedicated to Baal Hammon on the eastern side of the city, a highly localised ritual dating back possibly to the 3rd century BC (Gilhaus, 2021:170). Fewer than 40 of these contain any kind of written inscription, but 15 of these are inscribed with the names of Roman citizens in Latin, and several of these inscriptions lack cognomen. This implies that the dedicators were members of the Pagus which arrived at Thugga prior to the common adoption of cognomen by large portions of the population (Gilhaus, 2021:172). Rather than the local inhabitants placing a thin Roman veneer over pre-existing indigenous gods or abandoning local traditions entirely in favour of imposed Roman ones, it is clear that some members of the *Pagus* were participating in highly localised spirituality. Within this complex religious environment, already defined by dominant pre-existing chthonic traditions, the establishment of the *Pagus* could have plausibly catalysed a reinterpretation of the Roman ancestor gods, the *Manes*, producing a measurable impact on the region's commemorative culture. 

Is it possible then, that the arrival of the *Pagus* to a region with pre-existing chthonic traditions was a catalyst for a reinterpretation of the Roman *Manes* and other ancestor deities, perhaps as powerful and influential spirits that had to be commemorated correctly, rather than as individuals? The secular qualities of the deceased become irrelevant if the objective is not immortal memory and individual legacy, but ritual correctness. If the Manes of Thugga were unusually powerful and therefore ritually demanding, perhaps they had to be addressed correctly and rigidly, and this explains the heavy repeated use of formulae, and the almost omnipresent inclusion of reference to the deceased’s piety. To investigate the corpus further I extracted all longer (60-character) funerary inscriptions that abandoned the typical formula altogether. There were only 28 inscriptions of this kind (just 1.37% of the corpus), some of which were highly revealing.

### EDCS-69200511

**Latin**

DMS LICINIA M FIL ATTICA P V ANN LII HSE 
QVOD TIBI NON OLIM MATER VENERANDA SEPVLCRVM FECIMVS ARTIFICIS CAVSA FVERE MORAE

**English**

*Sacred to the Spirits of the Dead / Licinia, daughter of Marcus / Attica, dutiful, lived 52 years / here she lies / because we once / made a venerable tomb for you, mother / delays were the cause of the workmen's fault*

The first half of this inscription adheres perfectly to Thugga's dominant formula opening with DMS, certifying piety and age, closing with H.S.E. The second half is frankly bizarre. The dedicator directly addresses the deceased, attributing the monument's belated erection to the craftsman. That this explanation was considered worth carving in stone indicates the delay caused the dedicator sufficient anxiety to publicly petition his deceased mother with an excuse. Further attention should be paid to the specific word choice employed. *Veneranda* means to worship/venerate and in the gerundive case, it explicitly implies religious obligation to his mother. This kind of petition only makes sense if the Manes are conceived as powerful, and potentially malevolent entities if ritual obligations are not upheld, a conception not alien in the Roman world (Hope, 2009:103) In this case the obligation is evidently correct and timely commemoration. This inscription almost explicitly reveals a portion of the cosmological rationale that underpins the funerary epigraphy of Thugga and is not the only one that does so. 

### EDCS: 16201276

**Latin**

DMS Q IVLIVS CASTVS PVA LXXV PATER PARAVI SOLVM CONSTITVI MEIS IS EGI ANNOS QVINQVE ET SEPTVAGINTA SVPSTITVI NATVM QVI COLAT SEMPER LAREM HSE

**English**

*Sacred to the Spirits of the Dead. Quintus Julius Castus, lived piously 75 years. I, the father, prepared the ground, established it for my family. I lived seventy five years, survived by a son who always tends the Lar. Here he lies.*

This is a particularly striking example: *supstitui natum qui colat semper Larem*,"I have raised up / appointed a son whose purpose is to worship the Larem forever" (Grimwade, pers. comm., 2026). The grammar supports the reading: substituere carries the sense of appointing a successor, while the subjunctive *colat* marks a relative clause of purpose ("so that he might tend"), so that the son's cultic role is framed as the very end for which he is raised

### EDCS-16800151

**Latin**

DMS M MATTIVS M FIL ARN VIATOR OPTATIANVS PVA LXX HSE PIO PARENTI STATVIT ARAM FILIVS MERITVMQ HONOREM SERO QVAMVIS REDDIDIT

**English**

*Sacred to the Spirits of the Dead / Marcus Mattius, son of Marcus / of Arnsen, Viator Optatianus, pious, lived 70 years / he lies here / his son set up an altar to his dutiful parent / and although late, rendered deserved honor*.

This inscription is similar to EDCS:69200511 in character, although avoids attributing blame to any external actor. It follows the typical formula of funerary epigraphy at Thugga, with the second half acknowledging a delay in commemoration. It is interesting to note that the dedicator describes it as an altar *“aram”* rather than a tomb, indicating a clear ritual function to the grave itself. This ritual function may relate to the omnipresent inclusion of Sacrum in DMS. The dedicator both acknowledges his lateness but also makes it clear that he has fulfilled his obligation, absolving himself of guilt. What was considered “late” for commemoration at Thugga is unknowable but is evidently a ritual violation of some kind. As these petitions appear to be directed at the dead, it appears the *Manes* are conceived of as capable of granting reprieve for such violations. Equally noteworthy, is the fact that these petitions are inscribed directly into the commemorative stone, perhaps implying a direct relationship between the monument and the *Manes*, to some extent as a medium of communication. In some traditions, the *Manes* or a “spirit” more generally was considered to reside within the tomb itself, and failure to erect one would result in a vengeful wandering spirit (Hope, 2009:102). In North Africa meals would often be ritually sacrificed as the gravesite, pointing to a clear ritual function to the tomb, some interpretation of which likely existed at Thugga. A common theme amongst other inscriptions that totally abandon the typical formula of Thugga, is that they were commissioned in advance by the deceased themselves. Probably one of the most significant outliers of this is a poem inscribed onto a mausoleum. 

### EDCS: 16201258

**Latin**

L IVLIVS CVPITIANVS MELLITVS HOC MSOLAEVM MIHI ET VIRIAE ROGATAE VXORI VTRISQVE NOBIS VIVIS POSTERITATIQVE NOSTRAE ET IN MEMORIAM CVPITI PATRIS ET VENVSTAE MATRIS OPTIMORVM PARENTIVM INSTITVI ET DEDIC IDEM FELIX CVM ESSET ANNORVM LX HOC OPVS SIBI ET ROGATAE CONIVGI ABSOLVTVM DEDICAVIT INSVPER VIXIT ANNIS VI FIVNT TOTIVS VITAE EIVS QVOAD VIXIT AN LXVI HIC SITVS EST IDEMQVE ROGATA CONIVX EIVS CVM ESSET ANNORVM XXXXVIIII HOC OPVS CVM MELLITO MARITO SVO DEDICAVIT INSVPER VIXIT ANNIS XXIIII FIVNT TOTIVS VITAE EIVS QVOAD VIXIT AN LXXIIII DETRAHE SERTA COMIS ET AMORVM OBLITA TVORVM TRISTIS INOPS CVRTA VESTE THALIA VENI NON MANVS IDALIA LASCIVIAT IMPROBA VIRGA NEC FLVAT ANTE TVOS LVCIDA PALLA PEDES IVLIVS HOC FECI MELLITVS QVI VOCOR OLIM CVPITO PATRI MATRI VENVSTAE MEAE ME POSVI CONIVGEMQ MEAM MIHI IVNCTA ROGATAM VT SIT IN AETERNVM CONDITA FAMA LOCI VIXIMVS AD SATIEM PIETATEM IMPLEVIMVS AMBO PRAESTITIMVS SVBOLEM FEMINEAM DVPLICEM VOS QVOQVE QVI LEGITIS VERSVS ET FACTA PROBATIS DISCITE SIC VESTROS MERITO SANCIRE PARENTES VT TE GLAFRIANE EXCOLEREM TITVLOSQ RELINQVAM VIVOS VT HOC FACEREM FATA DEDERE MIHI IVLIVS HOC PETO NVNC A TE DOMINATOR AVERNI CVM MORIAR MANIBVS VT IACEANT OSSA QVIETA MIHI

**English**

*Withdraw also the fees and forget the loves of your own, Thalia, sad and poor, come in a shortened garment, Let not the mischievous rod play wantonly in the Idalian hand, Nor let the shining robe flow before your feet, I, Julius Mellitus, who was once called, did this, I desire for my father, mother, and my lovely one, I placed myself and my wife joined to me, Rogata, So that the fame of the place may be founded forever, We lived to satisfaction, we both fulfilled piety, We produced a double female offspring, You also who read these verses and approve the deeds, Learn thus to duly honor your parents, That I, Glafrianus, might cultivate you and leave behind titles, As long as the fates granted me to do this while alive, I, Julius, now ask this from you, ruler of Avernus, when I die, may my bones lie quiet among the Manes*

This is most remarkable, employing a range of rhetorical devices of a high cultural register. Although it appears to depart entirely from the typical formula, there remain several notable parallels. Firstly, the dedicator explicitly states the piety of both himself and his wife. More so using ad satiem – “to satiety/fullness” as pietas is an obligation rather than a quality. Additionally, it finishes with a direct petition to the “ruler of Avernus” Dominator Averni, to grant him peaceful rest amongst the Manes. With Avernus being the entrance to the underworld (Clark, 1996:261), this is likely addressed to Pluto. This is a highly formal petition, with the dedicator refusing to call the god by name, and instead by his title. Additionally, it is worth noting what is absent.

It is reasonable to assume, that the dedicator of this inscription was educated, and wealthy, due to both the length of the inscription and his impressive cultural repertoire. Despite this, there is absolutely no reference to his secular life achievements, profession, or his wealth, instead he only affirms his completion of what he considers to be his earthly duty, before formally petitioning Pluto for acceptance amongst the Manes. Functionally speaking, this inscription is not as significant of an outlier as it appears to be at first glance. Additionally, though more speculative,  “*discite sic vestros merito sancire parentes*” is a religious invocation or "learn thus to duly honour your parents" is potentially a religious invocation, encouraging the reader to uphold their own spiritual obligations to their deceased ancestors. The use of *Sancire prior* to parentes may hint at this, Sancire meaning to make sacred, to ratify or to consecrate. This conception may also be supported by the prior noted dominance of FIL and other markers of descent in Thugga’s funerary epigraphy. Roman inheritance law typically stipulated who was responsible for worshiping the Manes of the deceased, making lineage important in determining who owes ritual obligation to whom (King, 2022:9). Of the other 26 inscriptions that abandon the typical formula altogether, nearly all are highly emotive, or commissioned in advance by the deceased.

This indicates that the people of Thugga had agency and could depart from typical commemorative conventions if they wished, although evidently rarely did. A particularly striking example is EDCS:16201258 supstitui natum qui colat semper Larem or “I have raised up/appointed a son whose purpose is to worship the Lares forever”. This directly implies that the continued worship of the Lares is the core purpose of raising his son (Grimwade, personal communication, 2026).
It is typically thought that lasting memory was the core objective of commemoration across the Roman world (Hope, 2009:39). Epitaphs rarely dealt with spiritual or philosophical concerns and instead frequently sought to capture the identity of the individual (Hope, 1997:113 and 2009:37). This, however, does not appear to be true at Thugga. Further contradicting this conception is the existence of near identical gravestones, of different people with the same name at Thugga.

### Thuggan Euergertism and Religious Life
Euergetism was another avenue by which the affluent could secure lasting memory. This desire to leave a legacy inspired public projects and civic donations across the empire, usually accompanied by a lengthy inscription honouring the dedicator (Hope, 2009:34). Thugga was no stranger to euergetism, with the notable Gabinni family alone responsible for constructing 4 of the city’s temples (Fentress, 2021:698). However, the city’s non-religious public buildings lack dedications altogether, with the benefactors choosing to remain anonymous (Fentress, 2021:969), something which is extremely unusual in the Roman world (Hope, 2009:34). The temples that are accompanied by dedications, appear to be semi-private in nature, typically accessed by inconspicuous entrances from the street (Fentress, 2021:969), unusually they are scattered through residential areas, and appear to have been deeply personal to the dedicators (Fentress, 2021:969). 

The large number of these temples implies a highly religious character of the town, with 25 temples of this style identified thus far (Fentress, 2021:698). Acts of religious euergetism may have been essential in accessing prestigious priesthoods (Bertolazzi, 2016:86), reference to these priesthoods are some of the only personal identifiers that exist in Thugga’s funerary epigraphy. In a highly religious town, such priesthoods may have represented the apex of public life. Notably at Thugga, women were both eligible for such roles and actively participated in religious euergetism often holding powerful priesthoods (Bertolazzi, 2016:105). In the 3rd century the amount of money spent by women on such projects is almost equal to that of men (Bertolazzi, 2016:104). This may all be conducive to an atypical value system, which places little to no emphasis on individual legacy and identity after death. 

Based on the available lines of evidence it appears that the overwhelming formulaic standardisation of Thugga's funerary corpus possibly reflects a localised reinterpretation of the Manes as an active divine community potentially with agency over the admission of the dead into the Manes. Central to this interpretation is pietas, a concept worth examining carefully, as it is by far the most dominant and almost the only recurring trait across the corpus. Rather than simply representing religious devotion as the modern word piety does, pietas refers to adherence and participation in reciprocal social and religious relationships, often conceived spiritually as a collective relationship between an entire community and the gods (King, 2020:45). As evidenced explicitly by EDCS-16201258, completion of pietas at Thugga appears to have been intimately related to the afterlife in some way.  *Pietatem implevimus*, "we fulfilled pietas," its inclusion on nearly every epitaph implies that completion of this duty may have been a prerequisite to joining the *Manes* after death or related to the cthonic world in some capacity.

In this interpretation, the epigraphic monument functioned not just as an individual commemoration but as a component of a ritual, with the almost universal inclusion of pius/pia vixit certifying the deceased's diligent completion of pietas as a qualification for admission. In this context standardisation was functionally necessary because correct deployment of formulae was part of this ritual. Beyond this it is difficult to construct a fully evidence-based hypothesis. Whether the dedicator of EDCS-69200511 was attempting to placate the potentially vengeful spirit of his mother, or to guarantee her acceptance into the Manes by explaining away his belated commemoration, (perhaps both) is ultimately unknowable. What the epigraphic evidence does establish is that at Thugga, correct and timely commemoration was understood as an obligation with real spiritual consequences, a conclusion supported by the formulaic uniformity of the corpus, the emphasis on pietas, and the anxiety of those late in commemoration.  


## Belief Clusters
When discussing variation in interpretation of the Manes King coins the term “belief clusters” to explain reoccurring general overlaps in some conceptions pertaining to a deity (such as the Manes), albeit not universally (King, 2020:87). It is possible then, that the epigraphic habit of Thugga is a preservation of one of these belief clusters. This does not require a radical redefinition of the Manes. Conceptions of the Manes as powerful and requiring ritually specific worship existed throughout the Roman world (King, 2020:10). A localist interpretation of the Roman Manes built around a conception of ritual correctness related to the completion of obligatory pietas at Thugga is entirely coherent in a region with pre-existing chthonic traditions and a religious world defined by non-dogmatism. The Manes appear to have been widely considered to be “minor” deities with belief in the afterlife tenuous (Hope, 2009:39). In the highly religious town of Thugga (Fentress and Dufton, 2022:193), it appears that this was not the case, to the extent that ritual correctness and adherence was valued above secular individual legacy and memory.

The *Manes* were ancestor gods often thought to reside in the underworld (but in some traditions the sky too) who had a variety of powers that could be asserted upon the living. This ranged from assisting armies in battle, extending life and even witnessing oaths (King, 2020:8-63). There was a vast number of differing co-existing interpretations of the *Manes* and their associated powers, throughout the Roman world (King, 2020:62). There has been some debate as to whether they were invoked as the collective dead, but there are examples of individual ancestor worship too (King, 2020:35). Throughout the world the dead are often assigned power and agency (Pearson, 1993:203) and the Roman world was no exception to this. Death creates a crisis that is frequently understood within a cosmological framework, and resolved with ritual (Nilsson Stutz, 2015:4). The non-dogmatic nature of Greco-Roman spirituality makes it difficult to ascertain any universal characteristic of this cosmology or about the nature of the Manes beyond basic conceptions, such as the fact that they exist and possess some degree of power. Even this was not necessarily universal, with Epicureans questioning the value of worshipping the Manes altogether, albeit without questioning their existence (King, 2020:27). It is impossible to categorize and summarize all Roman attitudes towards death and belief in the afterlife (Hope, 1997:113), and there was almost unlimited room for variation. 

This interpretation does not require a radical interpretation of Roman cosmology. It simply requires a differing value system, one that places the worship of cthonic entities such as the *Lares* and *Manes* and adherence to spirtual obligations more generally above the desire for personal eternal memory. This fits Kings definition of a belief-cluster, most of these clusters are likely lost to time. The funerary record likely being one the few parts of the preservation record with the ability to encode evidence for such clusters. In Thugga's case it is not the product of direct pre-Roman continuity, or the neat layering of "Roman" spirtuality over local spirituality, but of re-imagination and a re-intpreation of the old and new catalysed by the arrival of a *Pagus*. 

## Future Research
Based on the number of temples per capita Bennett correctly notes that Thugga’s epigraphic habit is not unique.  To confirm this the FCI, average inscription length and most frequent terms for these nearby sites were assessed. Due to the small size of the corpora assessment of TTR and Herdan’s C TTR is made redundant by the corpus size affect. The template dominant at Thugga extends to both Uchi Maius, and Thiburisicum Bure, with all three sites being significant outliers when compared to Lambaesis and Rome. Uchi Maius is located just 12km west of Thugga, and Thiburisicum Bure just 4.7km north-east. Of the three Uchi Maius is the most divergent due to the limited presence of PVA (19 times) in favour of P/PIVS/PIA vix/it. Thurburisicum Bure is almost indistinguishable from Thugga, although the latter maintains a stronger preference for PVA, which appears just 54 times or in 28% of total inscriptions compared to 51.1% at Thugga. These measurable differences in deployment of PVA further contradict the copybook hypothesis, and perhaps point to independent stone masons operating in each site.
 
A topic of future research could be to see if this belief cluster exists and what its geographic range is. What is clear however, is that the application of corpus-linguistic methodology to the study of epigraphy is viable. The only explicit kind of limitation in this research is the data itself. This methodology could be applied to any kind of ancient writing, from any time period. Future analysis could include an aggregate of curse tablets from separate sites, monumental inscriptions and beyond. Corpus-linguistic software can be used to “datamine” ancient sources of text, for key terms or phrases and their context in a convenient and accurate manner. With the potential of significant time reductions enabling researchers to dramatically increase the scale and scope of study, even with limited resources. Equally the implementation of AI tools for coding related tasks, makes it now possible to automate large parts of menial research related work, such as the assembly of large datasets.
