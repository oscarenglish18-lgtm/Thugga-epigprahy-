The funerary epigraphy of Thugga (Dougga), Roman North Africa, is characterised by an unusual reliance on a limited range of formulae. This dissertation investigates to which extent this is statistically real by applying a corpus linguistic methodology to assess standardisation metrics at Thugga, Uchi Maius, Thibursicum Bure, Lambaesis and Rome. This comparative analysis involves 6,128 funerary inscriptions drawn from the Epigraphik-Datenbank Clauss-Slaby (EDCS) and compiled with a custom-built Python tool. Multiple standardisation metrics and other data extracted from these corpora point to the existence of unique commemorative preferences at Thugga, potentially related to a local interpretation of the Manes and a conception of obligatory pietas. In doing so, this research sheds light on the complex and localised religious environment of Roman North Africa, in which pre-existing Numidian chthonic traditions and incoming Roman spiritual practices were actively renegotiated rather than simply replaced or assimilated. 

## Introduction

North Africa and the Maghreb possess a rich archaeological record, and the study of epigraphy has dominated scholarship in the region (Mattingly and Hitchner, 1995:166-169). The region has long existed as a cultural crossroad, which is reflected in its tradition of diverse commemorative practices (Stone and Stirling, 2007:7). This is a legacy that can be traced long before the Roman occupation. The indigenous Berber population seems to have inhabited the region from at least 7000 BCE (Brett and Fentress, 1996:12) and frequently interacted with the wider Mediterranean world, with tombs morphologically similar to those produced by the Berber people found in Sicily, potentially reflective of this interconnectivity (Brett and Fentress, 1996:16). The Berbers were not necessarily a single coherent culture or identity and encompassed a diverse variety of traditions (Mattingly, 2019:26). Several of these would synthesize into a number of kingdoms, collectively labelled as “Numidians” by the recently arrived Phoenicians (Brett and Fentress, 1996:4).

By the 9th-10th centuries BCE urbanism had emerged indigenously, even deep into the desert (Fentress and Dufton, 2022:174). Although this coincided with the foundation of Carthage, it appears that Mediterranean influences did not replace local cultural and spiritual practices, with a degree of continuity persisting (Mattingly, 2019:18). Nonetheless, Punic influence extended far beyond the political control of Carthage (Brett and Fentress, 1996:24) creating a complex cultural environment. Several of these local cultures produced elaborate funerary traditions (Brett and Fentress, 1996:24).  The transformation of the region into a Roman province was iterative, beginning after the destruction of Carthage and annexation of Punic territory in 146 BCE, a border being delineated between the new Roman province of Africa Vetus and the Numidian Kingdom 90km inland (Hobson, 2022:103). By 43 BCE, the remaining Numidian lands were annexed following a Caesarean victory in the civil war, as a consequence of Numidian support for Pompey. This year also saw the refoundation of Carthage with Roman colonists, alongside the arrival of Roman settler communities to many pre-existing towns (Hobson, 2022:103). One such town was Thugga, a former Numidian centre and presently one of the best-preserved examples of a small Roman town in North Africa (UNESCO, 2026). 

In February 2025 I visited Thugga myself whilst travelling in Tunisia and was struck by a pattern in its funerary epigraphy. Nearly every epitaph I examined consisted of the same small cluster of abbreviations, a name, and an age at death. Not only were they extremely short, many deployed virtually identical formulae. They were each devoid of any reference to profession, personal narrative or elaboration. A subsequent survey of the Epigraphik-Datenbank Clauss-Slaby (EDCS), an online repository holding over two thousand funerary inscriptions from the site, confirmed this impression at scale. Thugga's funerary corpus appeared abbreviated, rigid, and almost entirely stripped of individual secular or personal content. This raised an immediate question, whether and to what extent this pattern was statistically real, how it compared to other sites in Roman North Africa, and what the implications of this mean for commemorative culture in the region. This unusual profile has been noted before, Bennett identified Thugga's anomalous reliance on a restricted formulaic register in a macro-level analysis of over 100,000 Roman epitaphs as part of a PhD thesis, attributing it to copybook stonemasons, but explicitly called for more detailed future research (Bennett, 2020:194).

Rather than the copybook template hypothesis, this dissertation will argue that a synthesis of the evidence points to a deliberate commemorative preference for unusual formulae and standardisation at Thugga, likely related to a localist interpretation of the Roman Manes and a conception of ritual correctness related to the completion of obligatory pietas. To demonstrate this, the dissertation applies corpus-linguistic methodology and software to complete a comparative analysis of funerary epigraphy across five sites in Roman North Africa and Rome using 6128 unique inscriptions. These were drawn from the Epigraphik-Datenbank Clauss-Slaby (EDCS) and processed using EDCS CorpusBuilder, a custom-built Python tool developed for this research, which automates the cleaning, preparation and generation of epigraphic corpora. The results of this research shed light on the complicated bilateral religious environment of Roman North Africa which saw aspects of localist and Roman spirituality reinterpreted in a new imperial context. 
 



## Historiography, The Manes, Commemoration and Spirituality in Roman North Africa

Roman annexation did not simply impose a new religious order on North Africa. Evidence points to a process of reinterpretation in which pre-existing spiritual traditions and incoming Roman ones were renegotiated within precise local contexts. Spirituality and commemorative culture are typically interlinked and thus must be assessed together (Scarre, 2011:9). Additionally, it is important to address outdated interpretative frameworks that were once dominant in North African Roman scholarship which can ultimately be traced to colonial rule in the region. 

### Colonial Archaeology and Spirituality in Roman Africa 

The study of Roman North Africa was fundamentally shaped by the environment it emerged, largely as a legitimising tool for European colonial rulers. France and Italy sought to portray themselves as the inheritors of a Roman European legacy and used archaeology to draw explicit comparisons between themselves and the Roman Empire (Mattingly, 2024:760). One of the consequences of this historiographical agenda was the prioritisation of Latin Epigraphy (McCarty, 2024:8), something as conceived of as evidence of Roman cultural imposition on the region (Hitchner, 2022:4). On the other hand, the pre-Roman period was correspondingly neglected and remains understudied (Mattingly, 2024:761). This ultimately led to the development of an interpretive framework built around a concept of “Romanisation” as an assumed process in which either the pre-existing North Africa population was unilaterally assimilated into Roman culture, or in which a thin Roman veneer was placed over pre-existing traditions, evidenced by indigenous participation in the Latin language and in “Roman” cultural and religious life. (Hitchner, 2022:4 and McCarty, 2024:6). Underpinning this conception was a 3-way ethnic division of the population, into “Punic”, “African” and “Roman” categories, which were considered to be religiously and culturally coherent (Fentress, 1978:507). In reality, these categories only “exist in discourse” (McCarty, 2022:286), identity in Roman North Africa was constantly renegotiated, and neat lines of delineation along an ethnic basis cannot be drawn in a region as diverse as Roman North Africa (McCarty, 2022:286). In Roman North Africa, identity depended less on cultural or ethnic origin, and more on status, wealth, location, employment and religion (Mattingly, 2010:242). 

Following decolonisation and the emergence of post-colonial scholarship, the concept of Romanisation was initially flipped rather than dismantled. Marcel Benabou’s La résistance africaine à la Romanisation (1976) constructed a counter-narrative of eternal indigenous resistance to the Roman occupation (Hitchner, 2022:6), but this still posited the existence of monolithic African and Roman culture, existing in binary opposition to one another. In reality North Africa was highly religiously and culturally diverse, host to a patchwork of variable spiritual traditions whose patterns should not be explained by cultural difference but examined within precise local contexts (McCarty, 2022:286). Although this diversity persisted into the Roman period, attempts to explain the ‘peculiarities’ of religion in Roman North Africa, such as the unusual popularity of Saturn by assuming that they represent the continued worship of a pre-Roman god (in the case of Saturn-Baal Hammon) (Cadotte, 2006) are inaccurate (McCarty, 2022:286). North Africa was highly interconnected, to the extent that Roman rule itself was preceded by the arrival of Greco-Roman gods, with an interpretation of Pluto arriving in the Punic period (Cadotte, 2006). This notion of direct continuity is problematic, as in the Roman world, religious and non-religious behaviour were not clearly distinguished (Hingley, 2011:747). The idea of straightforward religious continuity in the context of radical disruption caused by Roman annexation, the destruction of Carthage and the introduction of a Roman citizen Pagus to at least 8 pre-existing towns (Fentress, 2021:965) is not plausible. This radical disruption ultimately led to a reinterpretation and the reshaping of old gods and practices, adapted for new and changing social frameworks (McCarty, 2022:289-296).

Although the Romans were practitioners of Interpretatio Romana, a process in which a foreign god would be integrated or identified with a pre-existing member of the Roman pantheon, this was rarely straightforward transliteration (Ando, 2008:43). The process of this identification was not standardised or regulated and instead was one part of what was typically a complex cultural negotiation and dialogue (Ando, 2008:52). The non-dogmatic nature of Greco-Roman religion allowed for and accommodated simultaneous conflicting interpretations of the same gods. There was often substantial overlap in the jurisdiction of different gods, and complicating things further was the fact that the same god would often be known by multiple different names (Ferrero and Tridimas, 2018:162). This non-dogmatic nature allows for theoretically enormous diversity in practice and conception (King, 2020:86). Additionally, this non-dogmatism allows for the reimagination or reinvention of aspects of the divine. Such a process may have occurred in early Roman North Africa, as indicated by an unusual boom in Tophet construction that appears to have occurred in the aftermath of the destruction of Carthage (McCarty, 2017:398). Unlike typical Carthaginian examples however, the crescent that decorates these Tophets is inverted, and contents used for molk burnings differ significantly, with one example consisting entirely of plant matter, rather than infants or animal sacrifices (McCarty, 2017:405). This is not an example of Carthaginian continuity, but rather reimagination (McCarty, 2017).  This process of re-interpretation was assisted by the Roman tradition of assuming the existence of place-based, unknown deities to the extent that they would petition them generically as “gods of this place” when unsure (Ando, 2008:128). Roman settler participation in pre-existing local traditions is evidenced (Gilhaus, 2021:172).  The religious environment of Early Roman North Africa was one defined by both the introduction of the new and a reimagination of the old, not one of direct continuity nor replacement. 


North Africa presented a patchwork of loosely related deities and place-based spirituality that did not constitute a codified religion, and this was not nullified by Roman annexation (Brett and Fentress, 1996:35-49). There appears to have also been a level of geographic divergence between the coastal belt and interior, defined by the dominance of chthonic cults in former Numidian lands (Mattingly, 2024:773) in which Thugga itself lies. Evidence for this chthonic tradition can be found in the city, which hosted a temple to the deified Numidian king Massinissa (McCarty, 2022:289). An outdated model of Romanization has frequently been applied to Thugga, particularly surrounding the arrival of the Pagus, but also when assessing its pre-Roman “Punic” or “African” character. For example, the existence of Sufetes at Thugga were assumed to be a preservation or direct adoption of the Carthaginian institution, in reality it appears to have been a local reinvention based on a distant memory with a significantly different set of functions and responsibilities (McCarty, 2017:412). Although it is true that the Pagus remodelled large parts of the forum and dismantled pre-existing temples (Fentress, 2021:965) there is evidence of a more complex religious relationship lurking beneath as well. For example, the people of Thugga continued to deposit stelae dedicated to Baal Hammon on the eastern side of the city, a highly localised ritual dating back possibly to the 3rd century BC (Gilhaus, 2021:170). Fewer than 40 of these contain any kind of written inscription, but 15 of these are inscribed with the names of Roman citizens in Latin, and several of these inscriptions lack cognomen. This implies that the dedicators were members of the Pagus which arrived at Thugga prior to the common adoption of cognomen by large portions of the population (Gilhaus, 2021:172). Rather than the local inhabitants placing a thin Roman veneer over pre-existing indigenous gods or abandoning local traditions entirely in favour of imposed Roman ones, it is clear that some members of the Pagus were participating in highly localised spirituality. Within this complex religious environment, already defined by dominant pre-existing chthonic traditions, the establishment of the Pagus could have plausibly catalysed a reinterpretation of the Roman ancestor gods, the Manes, producing a measurable impact on the region's commemorative culture.

### Roman Commemoration and the Manes

The Manes were ancestor gods often thought to reside in the underworld (but in some traditions the sky too) who had a variety of powers that could be asserted upon the living. This ranged from assisting armies in battle, extending life and even witnessing oaths (King, 2020:8-63). There was a vast number of differing co-existing interpretations of the Manes and their associated powers, throughout the Roman world (King, 2020:62). There has been some debate as to whether they were invoked as the collective dead, but there are examples of individual ancestor worship too (King, 2020:35). Throughout the world the dead are often assigned power and agency (Pearson, 1993:203) and the Roman world was no exception to this. Death creates a crisis that is frequently understood within a cosmological framework, and resolved with ritual (Nilsson Stutz, 2015:4). The non-dogmatic nature of Greco-Roman spirituality makes it difficult to ascertain any universal characteristic of this cosmology or about the nature of the Manes beyond basic conceptions, such as the fact that they exist and possess some degree of power. Even this was not necessarily universal, with Epicureans questioning the value of worshipping the Manes altogether, albeit without questioning their existence (King, 2020:27). It is impossible to categorize and summarize all Roman attitudes towards death and belief in the afterlife (Hope, 1997:113), and there was almost unlimited room for variation. 

The Roman Calendar included two festivals relating to the dead. Lemuria, which was celebrated on the 9th, 11th and 13th of May and Parentalia, which was held 13th-21st of February (Hope, 2009:99). Lemuria was based around a ghostly midnight ritual in which an offering of black beans would be left for the Lemures - the malignant spirits of dead ancestors, before then being ritually exorcised from the home (Hope, 2009:99). It seems to have been reasonably common to conceive of the dead as potentially malignant (Hope, 2009:99). The division of the realm of the living from the realm of the dead appears to have been important, and upholding this division may have been a function of Lemuria. Like all other aspects of the Roman religion, conceptions varied. In addition to the Lemures there were also the Lares which were household ancestor gods of a less hostile nature, however sometimes it appears no distinction between Lares and Manes was made (King, 2020:63). Parentalia was a decidedly less ghostly ritual in which people would gather to remember deceased loved ones. Typically various offerings would be left by the grave, perhaps implying that the dead were to some extent capable of receiving sustenance (Hope, 2009:102). The actual significance of these festivals, and adherence to them is largely unknowable. The most detailed explanation for these festivals was produced by Ovid writing under the Augustan regime, thus he may have reinvented the significance of these festivals to appeal to the emperor’s emphasis on piety (Hope, 2009:99). In North Africa textual sources point to a differing interpretation of Parentalia in which a portion of a meal would be sacrificed to the dead. This ritual persisted into the Christian period as evidenced by clerical criticism of the practice (Rebillard, 2015). This is yet another example of a re-interpreted tradition in North Africa, perhaps related to a prior memory and understanding of ancestor deities. 

Although numerous literary works detail an afterlife and posit the existence of a hell-like realm called Tartarus and the correspondingly heaven-like kingdom of Elysium, most of these works have strong moral undercurrents, and the extent to which these stories were interpreted literally is uncertain (Hope, 2009:114). Complicating this is the fact that most epitaphs are short and lack explicit reference to the afterlife altogether (Hope, 2009:115).  Across the empire tombs rarely dealt with spiritual or philosophical concerns about the soul (Hope, 2009:37), instead it appears that the primary objective was individual memory and legacy, perhaps as a result of a tenuous belief in the afterlife (Hope, 2009:39). This typically related to a Roman conception of “immortality through memory” or praemium aeternitatis, with Pliny the Younger going as far as to state in a letter that pursuit of this eternal glory is the only good reason to practice virtue (Naccarato, 2025:242). However, the act of commemoration is inherently personal, and attitudes differed, with the Roman general Frontinius forbidding the erection of a monument to himself as he felt that someone’s memory would last only if deserved (Naccarato, 2025:243). Despite the lack of overt spiritual content in epitaphs, the opening of DM, short for Dis Manibus, appears to have become ubiquitous in the 1st century AD (Hope, 2009:115). Dis Manibus is an invocation of the Manes meaning “to the spirits of the departed”, it is uncertain to what extent this was reflective of belief, or perhaps it became an unquestioned convention (Hope, 2009:115). Once again, evidently a somewhat differing interpretation of this convention existed in North Africa, with the addition of an S for Sacrum “Dis Manibus Sacrum”. The cause for this addition remains unclear, although it is clearly a religious extension of the phrase, Sacrum meaning sacred or in full, “to the sacred spirits of the departed”. This perhaps, is another hint of a unique interpretation of the Manes, perhaps as a more spiritually significant force. At Thugga, “DMS” is totally dominant, “DM” appears just 8 times in a corpus of 2048 inscriptions, whereas “DMS” appears 1582 times. Across all of North Africa, DMS appears in 51% of inscriptions, in the Western provinces it appears in just 1% of all epitaphs, and in Rome 3% (Bennett, 2021:163). The preference for DMS is therefore not unique to Thugga, but it is especially dominant, even in the context of the wider region. 

It is not possible to divorce a funerary record from the cosmological or spiritual rationale that underpins it (Scarre, 2011:9). Although universal generalisations about death rituals are indefensible, they are typically built around a concept of liminality, in which ritual is a necessary component in transforming the individual from the world of the living to the world of the dead (Pearson, 1993:204). The spiritual is frequently actualised in material form (Pearson, 1993:203), an example of which can be the funerary record. It is therefore impossible to assess Thugga’s unusual commemorative record, without also assessing the potential spiritual rationale and context that may underpin it. First however, it is necessary to quantify to exactly what extent Thugga’s funerary record is unusual in the Roman world. 


## Digital Epigraphy and Methodology

Corpus linguistics is a quantitative methodology used within the field of linguistics, in which computer software is used to identify and count linguistic features, often with the objective of producing statistical models (McEnery and Wilson, 2001). It can be applied to many, if not all, aspects of language (Hardie, 2016:502), and therefore funerary epigraphy. It can be used to measure anything from the prevalence of certain formulae, broader patterns of collocation or to extract key relevant words and terms. A prerequisite for this, of course, is a dataset or corpus of machine-readable text. Producing such a dataset is logistically difficult and a significant constraint to conducting macro-level epigraphic research of this kind. The assembly of such a corpus usually requires painstaking manual labour, or the production of custom-built software tailored to the project (Hardie, 2016:510). This creates a technical barrier, however as I will go on to demonstrate this is now surmountable with the application of LLM coding. The development of corpus linguistics has been inextricably linked to the humanities. The first corpus was produced by the Jesuit priest and PhD student Roberto Busa whilst studying the writings of Thomas Aquinas, a project he finished in 1967 (McEnery and Wilson, 2001:20). Although it has been used in the study of knowledge production in the field of archaeology (Schwartz, 2024), large scale comparative studies of epigraphic data remain rare (Hermankova et al, 2022). 

## Quantitative and Digital Approaches in Epigraphy 

The study of epigraphy has been historically dominated by detailed analysis and reading rather than broad comparative approaches, although this has started to change (Hermankova et al, 2022). Increasingly, the macro-scale study of epigraphy has been employed by archaeologists in the study of ancient demography, economics, migrations and socio-political organisation (Schiedel, 2007:3). These kinds of computational studies have largely been beyond the capabilities of any individual researcher (until recently) (Hermankova et al, 2022).  Despite this, the study of digital epigraphy has seen rapid growth in recent years, enabled by continued technological development (Hermankova et al, 2024). This rapid development however, is still constrained by several endemic issues, beyond the mere technical challenge of assembling a dataset. A significant limitation is the fact that the study of epigraphy, and ancient writing in general is decentralised across multiple academic disciplines (Hermankova et al, 2024). Epigraphers, archaeologists, linguists and classicists all employ differing methodological and theoretical frameworks. This impairs cross-discipline cooperation and knowledge transfer, something that is becoming increasingly recognised (De Santis and Rossi, 2018:13). A similar degree of disjointedness also persists on the technical side of the field, although now nearly 600,000 Latin inscriptions have been digitised (Hermankova et al, 2024), this has occurred across multiple databases, each employing separate terminologies, metadata practices, and classification systems (Hermankova et al, 2024). Finally recent rapid technological development has outpaced theoretical development in academia, with no standardised approach existing the research opportunities provided by this methodology have not been fully exploited (Bodel, 2024:2). 
More specifically, corpus linguistics remains largely restricted to the discipline of linguistics itself. Although a recent paper, published in March of 2026 assessed a corpus of 1650 late Roman inscriptions using corpus linguistics, it did so through an explicitly sociolinguistic, rather than archaeological lens. Specifically, to assess the reduction of AE to A in Latin inscriptions between the fourth and sixth centuries (Papini, 2026:112) something associated with the later evolution of Latin into a proto-Romance language. Bennett’s work employed a similar methodology, however largely depending on excel regex searches rather than corpus linguistic software, to identify common formula and assess other metrics such as average characters per inscription (Bennett, 2020:64). However, Bennett’s extremely broad approach meant that beyond calculating average inscription length and identifying prevalence of certain formula via regex searches, a more detailed analysis was not possible


## Site Selection and Dating

### Thugga

Thugga is central to this research, previously identified as a possessing an unusual epigraphic habit. Not only does it possess an extremely rich and well-preserved epigraphic record, but also an interesting and unusual cultural history. The date of the foundation of Thugga is unknown, but it was first recorded in the 4th century BCE when it was likely largely self-governing (Fentress and Dufton, 2022:191). It was not conquered violently, but shortly after the annexation of Numidian lands Roman settlers arrived, establishing a Pagus (Hitchner, 2002:598). This led to the development of an unusual dual administrative system defined by the existence of Sufetes (Fentress and Dufton, 2022:191) operating in parallel to the Roman magistrates and Pagus, for over two centuries (Hitchner, 2002:599).

### Lambaesis 

Lambaesis was first founded as a Roman military fort in 81 CE, and quickly also developed into a civilian town (Wilson, 2016). It would become the headquarters for Legio III Augusta, the only legion stationed in the region and one that would eventually be staffed largely by local recruits by the time of the second century (Faure, 2022:164). Lambaesis like Thugga has an excellent preservation record, with over 1500 individual funerary inscriptions digitised on EDCS. Additionally, it was host to a mix of cultural influences. Initially, the garrison comprised foreign soldiers alongside local civilians, many of whom intermarried. Over time it became host to soldiers of foreign descent born in Africa, as well as a large number of locals. The multi-ethnic nature of the Roman Army in North Africa “cannot be doubted” (Faure, 2022 p.164). This makes it an interesting candidate for corpus linguistic research, especially in comparison to Thugga, a city exposed to significantly different influences.

### Rome

Rome was selected as a metropolitan “baseline”. I wished to compare the data from Lambaesis and Thugga, with a city from outside of North Africa. This was necessary to differentiate between trends local to North Africa, and the individual site. Rome was the most obvious choice for this, as an imperial capital with the most expansive epigraphic record in the Roman world, making it ideal for corpus design

### Thuburiscium Bure and Uchi Maius 

These are both smaller sites, located in close proximity to Thugga. They were selected as additional comparisons to assess whether patterns observed at Thugga extended to neighbouring communities as observed by Bennett, 2020:189.

### Dating

Accurate dating is difficult, as even inscriptions with dating metadata typically consist of broad approximations rather than specific dates, sometimes spanning four centuries (Hansen, 2021: 142). An initial attempt to construct century-by-century corpora proved unworkable for this reason, as the resulting datasets were nearly identical. Site history is therefore used as a proxy for estimated date ranges. At Thugga urban space began contracting by the fourth century, and the surrounding rural landscape entered rapid decline by the seventh (Von Rummel, 2016: 109). The Thugga corpus is accordingly assigned a first-to-fourth century date range, with the caveat that funerary inscription production likely peaked alongside general construction activity in the second and early third centuries (Cilliers, 2016:88). Lambaesis follows a comparable trajectory, effectively collapsing by the fifth century, with parts of the urban area converted to grave use as early as 350 AD (Roskam, 1996: 47). Both North African corpora are therefore broadly chronologically analogous, consisting predominantly of early and middle imperial material. This estimate is extended to both Uchi Maius and Thiburiscium Bure, both located within 12km of Thugga. To ensure comparability, the Roman corpus was restricted to inscriptions dated between 1–400 AD, which the higher proportion of precisely dated Roman inscriptions makes possible.

## 1.4 EDCS CorpusBuilder

The assembly of these corpora required the creation of bespoke software - CorpusBuilder which is a custom Python tool built around a text normalisation script. It strips unnecessary EDCS metadata and eliminates author expansions identifiable by their appearance within () and <> brackets while preserving restored letters within []. It identifies 29 common abbreviations frequent in Latin funerary epigraphy (Elliot, 1998), capitalises all text, and converts U to V throughout. The program consists of 1,698 lines of code and can generate funerary corpora from 10 sites, all North African except Rome, using source data extracted from EDCS in March 2025. Corpus generation is controlled through a Tkinter GUI, through which the researcher selects the data source, number of inscriptions, date range, inclusion or exclusion of Greek texts, and a damage threshold of inscriptions with 3 eligible characters or fewer are excluded. After generation, the programs calculate several corpora statistics to give the user a basic overview, these include total tokens, inscriptions, average tokens per inscription and length filter. The output is a UTF-8 encoded .txt file suitable for direct import into corpus analysis software such as AntConc.
Architecturally, CorpusBuilder is a single Python script (or, when compiled, a single executable) that must reside in a folder containing a data subfolder. Within data are .txt files named according to site (e.g. thugga.txt, lambaesis.txt, rome.txt), each containing the raw inscription data exported directly from EDCS in its original, unprocessed form.

After this is completed it begins normalisation operations. This  includes stripping most EDCS metadata, such as findspot, author, province, editor ect. It removes all expansions identified as within () and <> editorial mark ups. Restored letters within [] are preserved, U becomes V and then a preset list of 29 commonly recognised abbreviations into single tokens. This is to prevent splintering of single abbreviations such as DMS into D, M , S which would artificially inflate token count and hinder any kind of analysis. 

```python
ABBREVIATION_PATTERNS = {
    r"\bD\s+M\s+S\b": "DMS",
    r"\bH\s+S\s+E\b": "HSE",
    ...  # 29 total patterns
}

def clean_inscription_lines(lines: List[str], greek_only: bool = False) -> str:
    ...
    line = re.sub(r"\([^)]*\)", "", original)   # strip author expansions
    line = re.sub(r"<[^>]*>", "", line)          # strip editorial markup
    line = line.replace("[", "").replace("]", "") # strip restored-letter brackets
    line = re.sub(r"[uU]", "V", line)             # normalise U to V
    ...
    for pattern, repl in ABBREVIATION_PATTERNS.items():
        line = re.sub(pattern, repl, line)
```

```python
def passes_length_filter(text, len_min, len_max) -> bool:
    if len_min is None and len_max is None:
        return True
    n = count_nonspace_chars(text)
    if len_min is not None and n < len_min:
        return False
    if len_max is not None and n > len_max:
        return False
    return True
```
